```python
df_new_death = df[['iso_code', 'location', 'date', 'new_deaths_per_million']]
df_new_death['date'] = pd.to_datetime(df_new_death['date'],format='%Y-%m-%d')
```

```python
df_new_death
```

```python
aux = df_new_death.groupby(['location','iso_code',df_new_death.date.dt.year])['new_deaths_per_million'].sum()
aux = aux.reset_index()
aux = aux.rename(columns={ 'location': 'name', 'iso_code':'code','new_deaths_per_million':'pop' })
```

```python
aux
```

```python
new_death_2020 = aux[aux['date']==2020]
new_death_2020 = new_death_2020.drop('date', axis=1)
new_death_2020.to_csv('new_death_2020.csv', index=False)
```

```python
new_death_2021 = aux[aux['date']==2021]
new_death_2021 = new_death_2021.drop('date', axis=1)
new_death_2021.to_csv('new_death_2021.csv', index=False)
```

```python
new_death_2022 = aux[aux['date']==2022]
new_death_2022 = new_death_2022.drop('date', axis=1)
new_death_2022.to_csv('new_death_2022.csv', index=False)
```

```python
new_death_2023 = aux[aux['date']==2023]
new_death_2023 = new_death_2023.drop('date', axis=1)
new_death_2023.to_csv('new_death_2023.csv', index=False)
```

```python
df_poverty = df[['location', 'date', 'extreme_poverty']]
df_poverty['date'] = pd.to_datetime(df_poverty['date'],format='%Y-%m-%d')
auxP = df_poverty.groupby(['location',df_poverty.date.dt.year])['extreme_poverty'].mean()
auxP = auxP.reset_index()
df_poverty_2023 = auxP[auxP['date']==2023]
df_poverty_2023 = df_poverty_2023.dropna();
df_poverty_2023 = df_poverty_2023.drop('date', axis=1)
df_poverty_2023.to_csv('df_poverty_2023.csv', index=False)
```

```python
df_clean = df[['location', 'date', 'handwashing_facilities',]]
df_clean['date'] = pd.to_datetime(df_clean['date'],format='%Y-%m-%d')
auxP = df_clean.groupby(['location',df_clean.date.dt.year])['handwashing_facilities'].mean()
auxP = auxP.reset_index()
df_clean_2023 = auxP[auxP['date']==2023]
df_clean_2023 = df_clean_2023.dropna();
df_clean_2023 = df_clean_2023.drop('date', axis=1)
df_clean_2023.to_csv('df_clean_2023.csv', index=False)
```

```python
df_gdp = df[['location', 'date', 'gdp_per_capita',]]
df_gdp['date'] = pd.to_datetime(df_gdp['date'],format='%Y-%m-%d')
auxP = df_gdp.groupby(['location',df_gdp.date.dt.year])['gdp_per_capita'].mean()
auxP = auxP.reset_index()
df_gdp_2023 = auxP[auxP['date']==2023]
df_gdp_2023 = df_gdp_2023.dropna();
df_gdp_2023 = df_gdp_2023.drop('date', axis=1)
df_gdp_2023.to_csv('df_gdp_2023.csv', index=False)
```
